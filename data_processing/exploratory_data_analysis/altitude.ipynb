```python
import pandas as pd
import matplotlib.pyplot as plt


def read_csv(file_path):
    return pd.read_csv(file_path)

file_path = '../../data/datasets/rodrigues/corrected_flights.csv'

#df means data frame
df = read_csv(file_path)
```

```python
#group by flight number. .first() because we want to compare altitude which is the same for all the rows of each flight
df = df.groupby('flight').first()

print(len(df))
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df.index, df['altitude'], color='skyblue')

plt.xlabel('Flight')
plt.ylabel('Altitude (m)')
plt.title('Height of the First Row in Each Flight in m')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
plt.scatter(df.index, df['altitude'], color='skyblue', marker='o', s=10)
```

```python
altitude_counts = df['altitude'].value_counts().reset_index()
altitude_counts.columns = ['Altitude', 'Flight Count']

# Display the table
altitude_counts
```

```python
# Display flights with altitude 0
flights_with_altitude_0 = df[df['altitude'] == 0].reset_index()[['flight', 'altitude']]
print("Flights with Altitude 0:")
print(flights_with_altitude_0)

# Display flights with altitude 3
flights_with_altitude_3 = df[df['altitude'] == 3].reset_index()[['flight', 'altitude']]
print("\nFlights with Altitude 3:")
print(flights_with_altitude_3)
```
